```python
from threading import Thread
import time
```

```python
def carro(numero,velocidade):
    trajeto = 0
    while trajeto <= 100:
        trajeto+= velocidade
        time.sleep(0.5)
        print('Carro{}: {} \n'.format(numero,trajeto))
```

```python
carro1 = Thread(target=carro, args=[1,1])
carro2 = Thread(target=carro, args=[2,3])

carro1.start()
carro2.start()
```
